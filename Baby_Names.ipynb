```python
# Plotting number of sex/name babies as a function of year

def plotname(sex, name):
    data = all_years_indexed.loc[(sex, name)]
    
    plt.plot(data.index, data.values, label = name)
    plt.axis(xmin = 1880, xmax = 2022)
```

```python
# Combining several "plotname()" plots for given sex and list of names

def comparenames(sex, names):
    plt.figure(figsize = (12, 3.5))
    
    for name in names:
        plotname(sex, name)
        
    plt.legend()
```

```python
comparenames('M', ['Michael', 'John', 'David', 'Thomas'])
```

```python
comparenames('F', ['Elizabeth', 'Claire', 'Emily', 'Anna'])
```

```python
# Making a function to get top ten names for sex and year

def getyear(sex, year):
    return (all_years_byyears.loc[(sex, year)]
           .sort_values('number', ascending = False)
           .head(10)
           .reset_index()
           .name)
```

```python
# Use of the function to fetch top 10 names of year 1880 of gender 'M'

getyear('M', 1880)
```

```python
# Printing the table of the names for each year of gender 'M'

pd.DataFrame({year : getyear('M', year) for year in range(1880, 2023)})
```

```python
# Printing the table of the names for each year of gender 'F'

pd.DataFrame({year : getyear('F', year) for year in range(1880, 2023)})
```

```python
# Function to create plot of all top 10 names by sex

def plotname(sex, name):
    data = all_years.query('sex == @sex and name == @name')
    
    plt.plot(data.year, data.number, label = name)
    plt.axis(xmin = 1880, xmax = 2023)
```

```python
plt.figure(figsize=(12,5.5))

for name in getyear('F', 2022):
    plotname('F', name)

plt.legend()
```

```python
# Converting the object into list

list(getyear('F', 2022))
```

```python
# Total number across all years, group by sex and name

totals = all_years.groupby(['sex', 'name']).number.sum()
```

```python
totals
```

```python
# Split into male and female totals

male, female = totals.loc['M'], totals.loc['F']
```

```python
ratios = (totals.loc['M'] / totals.loc['F']).dropna()
```

```python
ratios
```

```python
# Names that fit unisex ratio constraint

unisex = ratios[(ratios > 0.5) & (ratios < 2)].index
```

```python
unisex
```

```python
common = (male.loc[unisex] + female.loc[unisex]).sort_values(ascending = False).head(10)
```

```python
common
```

```python
all_years_indexed = all_years.set_index(['sex', 'name', 'year']).sort_index()
```

```python
all_years_indexed
```

```python
plt.figure(figsize = (9,9))

for i, name in enumerate(common.index):
    plt.subplot(5, 2, i+1)
    
    plt.plot(all_years_indexed.loc['M', name], label = 'M')
    plt.plot(all_years_indexed.loc['F', name], label = 'F')
             
    plt.legend()
    plt.title(name)
             
plt.tight_layout()
```
